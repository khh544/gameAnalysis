### 기초데이터 분석(엔씨소프트 - 트위터 데이터)

```python
keyword = 'NC소프트 -is:retweet lang:ko'     # 자신이 검색하고 싶은 키워드 입력 
search = [] # 크롤링 결과 저장할 변수   

cnt = 1
# while True:
while(cnt <= 10):   # 1page 대상으로 크롤링
    tweets = api.search_tweets(keyword)
    for tweet in tweets:
        search.append(tweet)
        cnt += 1

print(cnt)

for i in range(cnt-1):
    print(search[i].text) # 저장된 tweets text 보기
    
# print(search[0].text) # 첫번째 text 보기 

#전체 문서 보기

data = {}   # 전체 문서 추가
i = 0       # 문서 번호
for tweet in search:
    data['date'] = tweet.created_at
    data['text'] = tweet.text   # text키에 text문서 저장
#     print(i, " : ", data)   # 문서번호 : 문서내용
    i += 1
```
